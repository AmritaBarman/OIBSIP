```python
#Data Cleaning
#checking for missing values
df.isnull().sum()
```

```python
#Handling missing values
#filling missing value of name and host_name with unknown name and unknown host
df.fillna({'name':'Unknown Name'}, inplace=True)
df.fillna({'host_name':'Unknown Host'}, inplace=True)
```

```python
#filling missing value of last_review and reviews_per_month with Never Reviewed and 0.0
df.fillna({'last_review':'Never Reviewed'}, inplace=True)
df.fillna({'reviews_per_month':0.0}, inplace=True)
```

```python
#final check for missing values
print(df.isnull().sum())
print("\nNo missing value left.")
```

```python
#removing duplicate values if any to prevent data duplication
df.drop_duplicates(inplace=True)
```

```python
#standardization 
# Minimum values in important non-negative fields
cols_to_check = ['price', 'reviews_per_month', 'number_of_reviews',
                 'minimum_nights', 'calculated_host_listings_count', 'availability_365']
print("Minimum values in key columns:")
for col in cols_to_check:
    print(f"{col}: {df[col].min()}")

#we remove any listings where price is zero, which is not valid as minimum nights is 1.
df = df[df['price'] > 0]
```

```python
df.info() #Data Integrity Check
```

```python
#outlier detection
#outliers of price
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Price IQR: {IQR}")
print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")

outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
print(f"Number of outliers: {len(outliers)}")

# Filter out price outliers
df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

```python
df.info() #Data Integrity check
```

```python
#outliers of calculated_host_listings_count
Q1 = df['calculated_host_listings_count'].quantile(0.25)
Q3 = df['calculated_host_listings_count'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Host Listing IQR: {IQR}")
print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")

outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
print(f"Number of outliers: {len(outliers)}")

# Filter out host listing outliers
df = df[(df['calculated_host_listings_count'] >= lower_bound) & (df['calculated_host_listings_count'] <= upper_bound)]
```
